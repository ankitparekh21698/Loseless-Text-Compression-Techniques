```python
import numpy
## We have considered we don't have any initial dictionary.
initial_dictionary = {}
final_dictionary = initial_dictionary
size = len(final_dictionary) ## initially zero size of the dictionary
data = "wabba-wabba-wabba-woo-woo-woo "

## Creating the dictionary
s=""
i=0
arr = []
while(i<len(data)):
    s=s+data[i]
    for k,v in final_dictionary.items():
        if(s==v):   ## checking if the substring is already present in the dictionary
            i=i+1   ## if already present then increasing the value of i so that substring increases
            if(i==len(data)):
               break     ## exiting the for loop if substring becomes equal to the whole input data
            s=s+data[i]    
    if(i==len(data)):
        break     ## exiting the while loop if substring becomes equal to the whole input data
    size = size + 1  ## increasing the size of the dictionary because substring not present in the dictionary
    final_dictionary[size] = s ## adding the substring to the dictionary
    print (" Appended Entry : ",s)
    s=s[:-1] ## removing the last character from the substring because LZW can introduce only 1 new character 
             ## in an entry if the entry is already present in the dictioanry. So,to send the index of the already
             ## added entry we need to remove one character from the current substring.
    for k,v in final_dictionary.items():
        if s==v:  
            arr.append(k)
            print (" Sent Index : ",k," for entry : ",s)
    s = "" ## reinitializing the substring

   
print (" 1. Input Data to be encoded : ",data)
print (" 2. Dictionary which is created : ")
print ("{:<8} {:<15}".format('Index','Entry'))
for k, v in final_dictionary.items():
    entry = v
    print ("{:<8} {:<15}".format(k,entry))
print (" 3. Encoded Data : ", arr)

##Decoding

decodedop=""
for i in range(len(arr)):
    decodedop += final_dictionary[arr[i]]
print (" 4. Decoded Data : ",decodedop)
```
