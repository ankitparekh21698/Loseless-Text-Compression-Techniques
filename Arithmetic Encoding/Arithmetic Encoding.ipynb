```python
import math
from fractions import Fraction
class color:
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
    BLUE = '\033[94m'
    RED = '\033[91m'
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    GREEN = '\033[92m'
data = 'GIANT'
character_count = len(data)
print(color.BOLD+color.UNDERLINE+color.GREEN+"Sequence                     "+color.END,data)
print(color.BOLD+color.UNDERLINE+color.GREEN+"Length of sequence           "+color.END,character_count)
freq = {}
for c in data:
    if c in freq:
        freq[c] += 1
    else:
        freq[c] = 1
print(color.BOLD+color.UNDERLINE+color.GREEN+"Frequency                    "+color.END,freq)
totalfreq=sum(freq.values())
probs = freq.copy()
cumprobs = freq.copy()
cum_sum = 0
h = 0
for x in probs:
    cumprobs[x] = Fraction(cum_sum,totalfreq)
    cum_sum = cum_sum + probs[x]
    probs[x] = Fraction(probs[x],totalfreq)
    h += probs[x]*math.log(1/probs[x])/math.log(2)
print(color.BOLD+color.UNDERLINE+color.GREEN+"Probabilities                "+color.END,probs)
print(color.BOLD+color.UNDERLINE+color.GREEN+"Cumulative Probabilities     "+color.END,cumprobs)
print(color.BOLD+color.UNDERLINE+color.GREEN+"Entropy                      "+color.END,h)


##COMPUTING TAG

frac=Fraction(0,1)
width=Fraction(1,1)
for c in data:
    frac = frac + cumprobs[c]*width
    width = width * probs[c]
print(color.BOLD+color.UNDERLINE+color.PURPLE+"Start Fraction               "+color.END,frac)
print(color.BOLD+color.UNDERLINE+color.PURPLE+"Width                        "+color.END,width)
print(color.BOLD+color.UNDERLINE+color.GREEN+"Fraction Range               "+color.END,frac," to ",frac+width)

##Considering midpoint of the tag interval to be tag value.
TAG_VALUE=(2*frac+width)/2
print(color.BOLD+color.UNDERLINE+color.GREEN+"Tag Value(considering midpt.)"+color.END,float(TAG_VALUE))
print(color.BOLD+color.UNDERLINE+color.GREEN+"Binary Rep. of Tag Value     "+color.END,bin(int(str(float(TAG_VALUE))[2:])))


##FINDING A VALUE IN THE TAG INTERVAL WHICH CAN BE REPRESENTED IN THE FORM OF N/2^B(BINARY FRACTION) where
## N is any integer and B is the no. of bits required to be sent corresponding to data.

def find_binary_fraction(input_start, input_end):
    output_fraction = Fraction(0, 1)
    output_denominator = 1

    while not (input_start <= output_fraction < input_end):
        output_numerator = 1 + ((input_start.numerator * output_denominator) // input_start.denominator)
        output_fraction = Fraction(output_numerator, output_denominator)
        output_denominator *= 2

    return output_fraction
output =find_binary_fraction(frac,frac+width)
print(color.BOLD+color.UNDERLINE+color.GREEN+"BINARY FRACTION              "+color.END,output)
print(color.BOLD+color.UNDERLINE+color.GREEN+"ENCODED NUMERATOR            "+color.END,str(bin(output.numerator))[2:])
print(color.BOLD+color.UNDERLINE+color.GREEN+"ENCODED DENOMINATOR          "+color.END,str(bin(output.denominator))[2:])

##WE NEED TO SEND PROBABILITY,CUMULATIVE PROBABILITY DICTIONARY AND LENGTH OF SEQUENCE TO RECEIVER ALONG WITH THE FRACTION


##DECODING THE FRACTION
s=''
ratio=cumprobs.copy()
for i in range(character_count):
    for c,cumprob in cumprobs.items():
        if TAG_VALUE>cumprob:
            decoded_char=c
            interval=probs[decoded_char]        
    s=s+decoded_char
    for x in cumprobs:
        cumprobs[x] =cumprobs[decoded_char] + ratio[x]*interval                

print(color.BOLD+color.UNDERLINE+color.PURPLE+"DECODED SEQUENCE             "+color.END,s)
```
